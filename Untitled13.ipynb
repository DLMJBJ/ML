```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,f1_score,roc_auc_score

loan_prediction_Y_train = pd.read_csv('C:\\Users\\JangBeomJun\\Downloads\\loan_prediction_Y_train.csv')
loan_prediction_X_train = pd.read_csv('C:\\Users\\JangBeomJun\\Downloads\\loan_prediction_X_train.csv')
loan_prediction_X_test = pd.read_csv('C:\\Users\\JangBeomJun\\Downloads\\loan_prediction_X_test.csv')

display(loan_prediction_Y_train.head())
display(loan_prediction_X_train.head())
display(loan_prediction_X_test.head())
```

```python
drop_col = ['Loan_ID']

tid = loan_prediction_X_test['Loan_ID']

y = loan_prediction_Y_train['Target']
```

```python
loan_prediction_X_test.isna().sum() #짜스
```

```python
x_test_drop = loan_prediction_X_test.drop(columns = drop_col)
x_train_drop = loan_prediction_X_train.drop(columns = drop_col)
```

```python
x_train_dum = pd.get_dummies(x_train_drop)
x_test_dum = pd.get_dummies(x_test_drop)
x_test_dum = x_test_dum[x_train_dum.columns]

x_train_dum.info()
x_test_dum.info()
```

```python
x_train,x_valid,y_train,y_valid = train_test_split(x_train_dum,y,test_size = 0.3,random_state = 12,stratify = y)
```

```python
rfc = RandomForestClassifier(random_state = 12)
gbc = GradientBoostingClassifier(random_state = 13)
dtc = DecisionTreeClassifier(random_state = 14)

rfc.fit(x_train,y_train)
gbc.fit(x_train,y_train)
dtc.fit(x_train,y_train)

rfc_pred = rfc.predict(x_valid)
gbc_pred = gbc.predict(x_valid)
dtc_pred = dtc.predict(x_valid)

valid_pred_proba = rfc.predict_proba(x_valid)[:,1]
valid_pred_proba1 = gbc.predict_proba(x_valid)[:,1]
valid_pred_proba2 = dtc.predict_proba(x_valid)[:,1]
```

```python
print('rfc_acc : ',accuracy_score(y_valid,rfc_pred))
print('rfc_f1 : ',f1_score(y_valid,rfc_pred, pos_label='Y'))
print('rfc_roc : ',roc_auc_score(y_valid,valid_pred_proba))

print('gbc_acc : ',accuracy_score(y_valid,gbc_pred))
print('gbc_f1 : ',f1_score(y_valid,gbc_pred, pos_label='Y'))
print('gbc_roc : ',roc_auc_score(y_valid,valid_pred_proba1))

print('dtc_acc : ',accuracy_score(y_valid,dtc_pred))
print('dtc_f1 : ',f1_score(y_valid,dtc_pred,pos_label='Y'))
print('dtc_roc : ',roc_auc_score(y_valid,valid_pred_proba2))
```

```python
test_pred = gbc.predict(x_test_dum) #gbc가  accuracy_score의 점수가 가장 높기 때문에, 더불어 f1 스코어는 0.5를 아득히 넘어가는 점수이기에 신뢰할 수 없는 지표라 판단되고, roc 스토어 또한 gbc가 가장 깔끔하기 때문에, 이전에도 이 3개의 모델을 써서 돌렸을 때 이게 더 높은 점수를 받았던것 같다.
exam1 = pd.DataFrame({'Loan_ID':tid,'Target' : test_pred})
exam1.to_csv('exam1.csv',mode='w')
```
