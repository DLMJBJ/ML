```python
import pandas as pd
import numpy as np
import datetime as dt

DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv'
df = pd.read_csv(DataUrl)
df.head(20)
```

```python
df = df.drop(labels = 'Unnamed: 0',axis=1)
np.shape(df)
```

```python
def func_mul(x):
    if x=='Unknown' :
        return 'N'
    elif x=='Less than $40K' :
        return 'a'
    elif x=='$40K - $60K' :
        return 'b'
    elif x=='$60K - $80K' :
        return 'c'
    elif x=='$80K - $120K' :
        return 'd'
    else :
        return 'e'





df['newIncome'] = list(map(func_mul,df['Income_Category']))
df['newIncome'].head()
```

```python
df['newIncom2'] = df['Income_Category'].apply(func_mul)
df['newIncom2'].head()
```

```python
def func_check(x):
    return str(int(int(x)/10)*10)


df['AgeState'] = list(map(func_check,df['Customer_Age']))
df['AgeState'].value_counts().sort_index()
```

```python
def func_Graduate(x):
    if 'Graduate' in x:
        return '1'
    else :
        return '0'
        


df['newEduLevel'] = list(map(func_Graduate,df['Education_Level']))
df['newEduLevel'].value_counts()
```

```python
def func_Credit(x):
    if int(x)>=4500:
        return '1'
    else :
        return '0'
        


df['newLimit'] = list(map(func_Credit,df['Credit_Limit']))
df['newLimit'].value_counts()
```

```python
def func_Two(df):
    if df['Marital_Status']=='Married':
        if df['Card_Category']=='Platinum' :
            return '1'
        else :
            return '0'
    else :
        return '0'

df['newState'] = df.apply(func_Two, axis=1)
df['newState'].value_counts()
```

```python
def func_Gender(df):
    if df['Gender']=='M':
        return 'male'
    else :
        return 'female'

df['Gender'] = df.apply(func_Gender, axis=1)
df['Gender'].value_counts()
```
