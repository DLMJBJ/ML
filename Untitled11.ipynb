```python
import pandas as pd
import numpy as np
import datetime as dt

studentscore_X_test = pd.read_csv('C://Users//JangBeomJun//Downloads//studentscore_X_test.csv')
studentscore_X_train = pd.read_csv('C://Users//JangBeomJun//Downloads//studentscore_X_train.csv')
studentscore_y_train = pd.read_csv('C://Users//JangBeomJun//Downloads//studentscore_y_train.csv')

display(studentscore_X_test.head())
display(studentscore_X_train.head())
display(studentscore_y_train.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score
```

```python
drop_col = ['StudentID']
tid = studentscore_X_test['StudentID']
```

```python
train_x_drop = studentscore_X_train.drop(columns = drop_col)
test_x_drop = studentscore_X_test.drop(columns = drop_col)
y_train_target = studentscore_y_train['G3']
```

```python
train_x_dum = pd.get_dummies(train_x_drop)
test_x_dum = pd.get_dummies(test_x_drop)[train_x_dum.columns]
```

```python
xtr,xt,ytr,yt = train_test_split(train_x_dum,y_train_target)
rf = RandomForestRegressor(random_state=10)
rf.fit(xtr,ytr)
y_validation_pred = rf.predict(xt)
```

```python
print('validation mse', mean_squared_error(yt,y_validation_pred)) #숫자가 크고, 
print('validation rmse', np.sqrt(mean_squared_error(yt,y_validation_pred)))  #그래서 루트를 하는
print('validation r2 score', r2_score(yt,y_validation_pred))
```

```python
test_pred = rf.predict(test_x_dum)
result = pd.DataFrame({'StudentID': tid,'G3' : test_pred})
result
```
