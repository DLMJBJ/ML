```python
drop_col = ['enrollee_id','city']
tid = test_x['enrollee_id']

train_x_drop = train_x.drop(columns = drop_col)

test_x_drop = test_x.drop(columns = drop_col)
y = train_y['target']
#널 데이터 처리
```

```python
train_x_dum = pd.get_dummies(train_x_drop)
test_x_dum = pd.get_dummies(test_x_drop)

test_x_dum = test_x_dum[train_x_dum.columns]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_valid,y_train,y_valid = train_test_split(train_x_dum,y,test_size = 0.3, random_state = 12, stratify = y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier


rfc = RandomForestClassifier(random_state = 8)
dtc = DecisionTreeClassifier(random_state = 8)
gbc = GradientBoostingClassifier(random_state = 8)

#학습 똭
rfc.fit(x_train,y_train)
dtc.fit(x_train,y_train)
gbc.fit(x_train,y_train)
```

```python
pred_valid = rfc.predict(x_valid)
pred_valid2 = dtc.predict(x_valid)
pred_valid3 = gbc.predict(x_valid)

from sklearn.metrics import f1_score,roc_auc_score

print('rfc_f1_score : \n',f1_score(y_valid,pred_valid))
print('rfc_roc_auc_score : \n',roc_auc_score(y_valid,pred_valid))
print()
print('dtc_f1_score : \n',f1_score(y_valid,pred_valid2))
print('dtc_roc_auc_score : \n',roc_auc_score(y_valid,pred_valid2))
print()
print('gbc_f1_score : \n',f1_score(y_valid,pred_valid3))
print('gbc_roc_auc_score : \n',roc_auc_score(y_valid,pred_valid3))
print()
print()
print('max_f1_score : ',max(f1_score(y_valid,pred_valid),f1_score(y_valid,pred_valid2),f1_score(y_valid,pred_valid3)))
print('max_roc_auc_score : ',max(roc_auc_score(y_valid,pred_valid),roc_auc_score(y_valid,pred_valid2),roc_auc_score(y_valid,pred_valid3)))

```
