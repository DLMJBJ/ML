```python
import pandas as pd
import numpy as np
import datetime as dt

DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv'
df = pd.read_csv(DataUrl)
df.head(20)
```

```python
df = df.drop(labels = 'Unnamed: 0',axis=1)
np.shape(df)
```

```python
def func_mul(x):
    if x=='Unknown' :
        return 'N'
    elif x=='Less than $40K' :
        return 'a'
    elif x=='$40K - $60K' :
        return 'b'
    elif x=='$60K - $80K' :
        return 'c'
    elif x=='$80K - $120K' :
        return 'd'
    else :
        return 'e'





df['newIncome'] = list(map(func_mul,df['Income_Category']))
df.head()
```

```python
df['newIncom'] = df['Income_Category'].apply(func_mul)
df.head()
```

```python
df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'])
df.info()
```

```python
def func_date(x):
    import datetime
    
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year,x.month,x.day))
    

    
df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(func_date)
df.head()
```
