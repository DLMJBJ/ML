```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression

TV_price_test = pd.read_csv('C:\\Users\\JangBeomJun\\Downloads\\TV_price_test.csv')
TV_price_train = pd.read_csv('C:\\Users\\JangBeomJun\\Downloads\\TV_price_train.csv')

display(TV_price_test.head())
display(TV_price_train.head())
```

```python
TV_price_test.isna().sum()
```

```python
# TV_price_train = TV_price_train.drop('판매가',axis=1)   이거 그냥 이용하자
```

```python
drop_col1 = ['ID']
drop_col2 = ['ID','판매가']
tid = TV_price_test['ID']
```

```python
y_train_target = TV_price_train['판매가']
train_x_drop = TV_price_train.drop(columns = drop_col2)
test_x_drop = TV_price_test.drop(columns = drop_col1)
target = TV_price_train['판매가']
```

```python
train_x_dum = pd.get_dummies(train_x_drop)
test_x_dum = pd.get_dummies(test_x_drop)
test_x_dum = test_x_dum[train_x_dum.columns]
```

```python
xtr,xt,ytr,yt = train_test_split(train_x_dum,y_train_target)

rf = RandomForestRegressor(random_state=10)
gb = GradientBoostingRegressor(random_state=10)
lr = LinearRegression()

rf.fit(xtr,ytr)
gb.fit(xtr,ytr)
lr.fit(xtr,ytr)

y_validation_pred = rf.predict(xt)
y_validation_pred1 = gb.predict(xt)
y_validation_pred2 = lr.predict(xt)
```

```python
print('validation rmse', np.sqrt(mean_squared_error(yt,y_validation_pred)))
print('validation1 rmse', np.sqrt(mean_squared_error(yt,y_validation_pred1)))
print('validation2rmse', np.sqrt(mean_squared_error(yt,y_validation_pred2)))

print('validation r2 score', r2_score(yt,y_validation_pred))
print('validation1 r2 score', r2_score(yt,y_validation_pred1))
print('validation2 r2 score', r2_score(yt,y_validation_pred2))
```

```python
test_pred = rf.predict(test_x_dum) # rmse 수치가 가장 낮고, r2 스코어가 가장 높기 때문에
exam2 = pd.DataFrame({'ID': tid,'판매가' : test_pred}) #이거 곱하기 10000하고 싶은데 시간이 없음. 람다 사용하면 되는데
print(exam2)
exam2.to_csv("exam2.csv", mode='w')
```
